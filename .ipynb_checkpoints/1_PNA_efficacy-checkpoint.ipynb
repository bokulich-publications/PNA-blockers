### Download the csv at level 5 and work with this one

```python
taxa_level5 = pd.read_csv('level-5(7).csv')

# Drop the pos ctrl
df_filtered = taxa_level5[~taxa_level5['Sample'].str.contains('pos', case=False, na=False)]
taxa_level5 = df_filtered
```

```python
# Identify analysis columns
analysis_columns = [col for col in taxa_level5.columns if col.startswith('d__Bacteria')]

# Normalize the values in the analysis columns to percentages
taxa_level5[analysis_columns] = taxa_level5[analysis_columns].div(taxa_level5[analysis_columns].sum(axis=1), axis=0) * 100

# Sort the DataFrame by 'mPNA' and 'pPNA'
df_sorted = taxa_level5.sort_values(by=['mPNA', 'pPNA'])

# Define colors
colors = []
for col in analysis_columns:
    if 'Chloroplast' in col:
        colors.append('green')
    elif 'Mitochondria' in col:
        colors.append('purple')
    else:
        colors.append('grey')
```

```python
df_sorted.head()
```

## Grouped Plot

```python
# specify columns we want to keep
additional_cols = ['PNA_composition', 'mPNA','pPNA', 'PNA_mix', 'Sample_ID']
cols_to_keep = additional_cols + analysis_columns

# subset to only keep interesting cols 
df_grouped = taxa_level5[cols_to_keep]

# cols_to_group: those NOT containing 'Chloroplast' or 'Mitochondria' 
cols_to_group = [col for col in analysis_columns if 'Chloroplast' not in col and 'Mitochondria' not in col]

# Sum the values of the filtered columns for each row and drop them 
df_grouped['Other_bacteria'] = df_grouped[cols_to_group].sum(axis=1)
df_grouped.drop(columns=cols_to_group, inplace=True)
df_grouped.rename(columns={'d__Bacteria;p__Proteobacteria;c__Alphaproteobacteria;o__Rickettsiales;f__Mitochondria':'Mitochondria', 'd__Bacteria;p__Cyanobacteria;c__Cyanobacteriia;o__Chloroplast;f__Chloroplast' : 'Chloroplast'}, inplace=True, errors='raise')  
```

```python
# Columns to convert to relative frequencies
columns_to_convert = ['Mitochondria', 'Chloroplast', 'Other_bacteria']

# Calculate the sum of the specified columns for each row
row_sums = df_grouped[columns_to_convert].sum(axis=1)

# Convert the values to relative frequencies (percentages)
df_grouped[columns_to_convert] = df_grouped[columns_to_convert].div(row_sums, axis=0) * 100
```

```python
df_grouped['PNA_composition'].unique()
```

### pPNA

```python
conditions = ['None0.00.0', 'pPNA0.01.25', 'pPNA0.01.0', 'pPNA0.00.75', 'pPNA0.00.5', 'pPNA0.00.25']
pPNA =  df_grouped[df_grouped['PNA_composition'].isin(conditions)]
pPNA.sort_values(by=['pPNA'], inplace=True)
```

```python
# Define the groups based on 'PNA_composition'
groups = ['None0.00.0', 'pPNA0.00.25', 'pPNA0.00.5', 'pPNA0.00.75', 'pPNA0.01.0', 'pPNA0.01.25']
data = [pPNA[pPNA['PNA_composition'] == group]['Chloroplast'] for group in groups]

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(*data)
print(f"ANOVA test statistic: {f_stat}")
print(f"P-value: {p_value}")

# If the p-value is significant, perform Tukey's HSD test for pairwise comparisons
if p_value < 0.05:
    print("Performing Tukey's HSD pairwise comparisons...")

    # Prepare data for Tukey's HSD (flatten the list and create a grouping variable)
    chloroplast_values = np.concatenate(data)
    group_labels = np.repeat(groups, [len(group) for group in data])

    # Perform Tukey's HSD test
    tukey_result = pairwise_tukeyhsd(chloroplast_values, group_labels, alpha=0.05)

    # Print results
    print(tukey_result.summary())
else:
    print("No significant differences found between groups. No pairwise comparison needed.")
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# CHLOROPLASTS 
# Box plot with transparency
box_plot = sns.boxplot(ax=axes[0], x='pPNA', y='Chloroplast', data=pPNA, fliersize=0, boxprops=dict(alpha=0.5), palette='Blues')
# Overlay strip plot
strip_plot = sns.stripplot(ax=axes[0], x='pPNA', y='Chloroplast', data=pPNA, jitter=False, color='navy', size=3)

# Calculate and annotate median values for Chloroplast
medians_chloroplast = pPNA.groupby('pPNA')['Chloroplast'].median()
for index, value in enumerate(medians_chloroplast):
    axes[0].text(index, value + 0.5, f'{value:.2f}', ha='center', va='bottom', color='black')

axes[0].set_xlabel('pPNA concentrations [ÂµM]')
axes[0].set_ylabel('Percentage of Chloroplast Contamination')


# MITOCHONDRIA  
# Box plot with transparency
box_plot = sns.boxplot(ax=axes[1], x='pPNA', y='Mitochondria', data=pPNA, fliersize=0, boxprops=dict(alpha=0.5), palette='Blues')
# Overlay strip plot
strip_plot = sns.stripplot(ax=axes[1], x='pPNA', y='Mitochondria', data=pPNA, jitter=False, color='navy', size=3)

# Calculate and annotate median values for Chloroplast
medians_chloroplast = pPNA.groupby('pPNA')['Mitochondria'].median()
for index, value in enumerate(medians_chloroplast):
    axes[1].text(index, value + 0.5, f'{value:.2f}', ha='center', va='bottom', color='black')

axes[1].set_xlabel('pPNA concentrations [ÂµM]')
axes[1].set_ylabel('Percentage of Mitochondria Contamination')


# Show the plot
plt.show()
```

# 2020 Taxa Barplot

These are the same grape must samples (385 total) sequenced in 2020 without PNA blockers. 

```python
must_2020 = pd.read_csv('2020-taxabarplot.csv')


taxa_of_interest = ['d__Bacteria;p__Cyanobacteria;c__Cyanobacteriia;o__Chloroplast;f__Chloroplast',
                    'd__Bacteria;p__Proteobacteria;c__Alphaproteobacteria;o__Rickettsiales;f__Mitochondria']
    
rename_dict = {'d__Bacteria;p__Cyanobacteria;c__Cyanobacteriia;o__Chloroplast;f__Chloroplast':'Chloroplast',
              'd__Bacteria;p__Proteobacteria;c__Alphaproteobacteria;o__Rickettsiales;f__Mitochondria':'Mitochondria'}


# rename the taxa of interest
must_2020 = must_2020.rename(columns=rename_dict)
# Sum up all columns that are not in the taxa_of_interest and add as 'Other'
must_2020['Other'] = must_2020.drop(columns=rename_dict.values()).sum(axis=1)
# Drop the original non-interest columns (already summed in 'Other')
must_2020 = must_2020[rename_dict.values()].join(must_2020['Other'])
# Calculate relative frequencies
must_2020 = must_2020.div(must_2020.sum(axis=1), axis=0) * 100

must_2020.head()
```
